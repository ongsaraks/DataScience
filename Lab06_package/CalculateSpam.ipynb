## Calculate Fitted Function Score for Detecting Spam:

```python
def f(x1, x2): return -1.2 + 0.3 * x1 + 0.2 * x2

p = f(5, 1)
p
```

### Calculate the Probability of Spam:

```python
import math
math.exp(p) / ( math.exp(p) +1 )
```

## Visualization of How Logistic Function changes values into Probabilities:

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(p): return math.exp(p) / ( math.exp(p) +1 )

range_min = -5
range_max = 5
x = [logistic(i) for i in np.arange(range_min,range_max,0.05)]
plt.plot(x)
plt.ylabel('Prob for f(x)')
_ = plt.title(f'Logistic function in Logistic Regression\np({min(x):.2f}) to p({max(x):.2f}).\nTransformed from f(x) in range {range_min:.2f} to {range_max:.2f}', fontsize=10)
plt.show()
print(x)
```
